```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

```

```python
# Read the CSV file into a pandas DataFrame
confirmed_cases_worldwide = pd.read_csv("datasets/confirmed_cases_worldwide.csv")
confirmed_cases_worldwide['date'] = pd.to_datetime(confirmed_cases_worldwide['date'])
```

```python
print(confirmed_cases_worldwide)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(confirmed_cases_worldwide['date'], confirmed_cases_worldwide['cum_cases'])  # Use 'date' and 'cum_cases'
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.title('Confirmed COVID-19 Cases Worldwide Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
confirmed_cases_china_vs_world = pd.read_csv("datasets/confirmed_cases_china_vs_world.csv")
confirmed_cases_china_vs_world['date'] = pd.to_datetime(confirmed_cases_china_vs_world['date'])

print(confirmed_cases_china_vs_world.head())
```

```python
plt.figure(figsize=(10, 6))
# Filter the DataFrame for China and world data separately
china_data = confirmed_cases_china_vs_world[confirmed_cases_china_vs_world['is_china'] == 'China']
world_data = confirmed_cases_china_vs_world[confirmed_cases_china_vs_world['is_china'] != 'China']

# Check if there is data for China and World
print("China data:")
print(china_data.head())
print("\nWorld data:")
print(world_data.head())
```

```python
plt.plot(china_data['date'], china_data['cum_cases'], label='China', color='blue')
plt.plot(world_data['date'], world_data['cum_cases'], label='World', color='green')
plt.xlabel('Date')
plt.ylabel('Cumulative confirmed cases (log scale)')
plt.title('Cumulative Confirmed COVID-19 Cases in China vs World')
plt.legend()
plt.xticks(rotation=45)
plt.yscale('log')  # Apply logarithmic scale to y-axis
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
confirmed_cases_top7_outside_china = pd.read_csv("datasets/confirmed_cases_top7_outside_china.csv")
confirmed_cases_top7_outside_china['date'] = pd.to_datetime(confirmed_cases_top7_outside_china['date'])

# Display a glimpse of the DataFrame
print(confirmed_cases_top7_outside_china.head())

# Draw a line plot of cum_cases vs. date, colored by country
plt.figure(figsize=(10, 6))
for country, data in confirmed_cases_top7_outside_china.groupby('country'):
    plt.plot(data['date'], data['cum_cases'], label=country)

plt.xlabel('Date')
plt.ylabel('Cumulative confirmed cases')
plt.title('Cumulative Confirmed COVID-19 Cases in Top 7 Countries Outside China')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
